# Task 02: Price Range Analysis

## ➣ Objective

- To analyze restaurant price ranges and study their relationship with aggregate ratings and rating colors.

## ➣ Tools Used

- Python  
- Pandas  
- Matplotlib  
- Seaborn  
- Google Colab  

## Step 1: Import Libraries & Load Dataset

- In this step, we import necessary libraries and load the dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Dataset.csv')
df.head()
```

## Step 2: Understanding Price Range Data

- Here, we explore the price range column in the dataset.

```python
df['Price range'].value_counts()
```

## Step 3: Most Common Price Range

- This step identifies the most common price range among restaurants.

```python
most_common_price = df['Price range'].mode()[0]

print("Most Common Price Range:", most_common_price)
```

## Step 4: Average Rating for Each Price Range

- This analysis calculates the average aggregate rating for each price range.

```python
avg_rating = df.groupby('Price range')['Aggregate rating'].mean()

print(avg_rating)
```

## Step 5: Visualization of Average Ratings by Price Range

- This chart compares average ratings across different price ranges.

```python
sns.barplot(
    x=avg_rating.index,
    y=avg_rating.values
)

plt.title("Average Rating by Price Range")
plt.ylabel("Average Rating")
plt.show()
```

## Step 6: Rating Colors for Different Price Ranges

- This step identifies the rating colors associated with different price ranges.

```python
rating_colors = df.groupby('Price range')['Rating color'].agg(lambda x: x.mode()[0])

print(rating_colors)
```

## Step 7: Highest Average Rating and Its Color

- This step identifies the price range with the highest average rating and its associated color.

```python
highest_rating_range = avg_rating.idxmax()
highest_rating_value = avg_rating.max()

highest_color = rating_colors[highest_rating_range]

print("Price Range with Highest Average Rating:", highest_rating_range)
print("Highest Average Rating:", highest_rating_value)
print("Associated Rating Color:", highest_color)
```

## ➣ Conclusion

- Identified the most common price range among restaurants  
- Calculated average ratings for each price range  
- Compared restaurant ratings across price categories  
- Identified rating colors associated with highest-rated price ranges  

This analysis helps understand the relationship between restaurant pricing and customer ratings.
